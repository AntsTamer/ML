```python
#1 MCM
a= int( input("Introduce el primer número:"))
b= int( input("Introduce el segundo número:"))

if (a>=b):
    min1=a
else:
    min1=b
while(1):
    if(min1%a==0)and (min1%b==0):
        print("El MCM de los números introducidos es: "+ str(min1))
        break
    min1=min1+1
```

```python
#2 MCD
a= int( input("Introduce el primer número:"))
b= int( input("Introduce el segundo número:"))

if (a>=b):
    min2=b
else:
    min2=a
while(1):
    if(a%min2==0)and (b%min2==0):
        print("El MCD de los números introducidos es: "+ str(min2))
        break
    min2=min2-1
```

```python
#3 Gráfico tarta
import matplotlib.pyplot as plt

x=[1,10,100,200]
labels =["Dato1", "Dato2","Dato3", "Dato4"]
colors=["c","b","g","y" ]
plt.pie(x, labels=labels,autopct="%.1f%%", colors=colors)

plt.show()

#Un diagrama de pastel se utiliza para visualizar facilmabete proporciones correspondientes a cada categoría.
```

```python
#4 Scatter plot
import matplotlib.pyplot as plt

x=[1,10,100,200]
y=[5,50,500,1000]
plt.scatter(x,y, color='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph1')
plt.show()

#Un gráfico de puntos se utiliza para referenciar datos cuantitativos de forma organizada.
```

```python
#5 Subplot

#Para mostrar distintas gráficas de forma paralela
#será necesario seguir los siguientes pasos.

#5.1. Cargar la librería siguiente
#con el siguiente cargamos y creamos acceso directo a la librería deseada.

import matplotlib.pyplot as plt

#A continuación cargamos distintas funciones matemáticas.
#Las 2 primeras nos facilitan dibujar nuestro eje x y las 3 últimas las funciones para el eje y.
from numpy import pi,linspace, sin, cos, tan

#Definición de las variables
x1=linspace(0,4*pi,100)
y1=sin(x1)
y2=-sin(x1)
y3=cos(x1)
y4=tan(x1)

#Creación de las gráficas
#Con el comando subplot tenemos la posibilidad de reducir el tamaño de la gráfica
#subplot(y,x,p) y reduce el eje y, x reduce el eje x y p determina la posición de dicha gráfica.
#Sin este comando las diferente nubes de puntos compartiriían el mismo espacio
plt.subplot(1,7,1)
plt.plot(x1,y1,"r-h")
plt.subplot(1,7,3)
plt.plot(x1,y2, "b-h")
plt.subplot(1,7,5)
plt.plot(x1,y3,"k")
plt.subplot(1,7,7)
plt.plot(x1,y4,"g")
plt.show()
```
