# Parte1

Lee archivo FicheroALeer.txt.

```python
#Abrir, cargar y printar archivo
with open ("FicheroALeer.txt", "r") as myfile:
    Fileloaded = myfile.read()
    print(Fileloaded)
```

Número de veces que aparece el artículo el en el archivo cargado.

```python
#Inicializacion de variables para los articulos
freq1=0
freq2=0
freq3=0
#Cuenta
freq1=Fileloaded.count(' El ')
freq2=Fileloaded.count(' el ')
freq3=freq1+freq2
#Printar cuenta
print(freq3)
```

Eliminación de tildes y paso de mayúscula a minúscula.

```python
myfile = open ('FicheroALeer.txt','r')

#Posicion de inicio de lectura en el archivo
myfile.seek(0)
#Declaracion de variable para la lectura de las lineas del archivo
textoinlines=myfile.readlines();

#Funcion para cambiar de mayusculas a minuscula
#Y eliminacion de tildes
def quitartildes(t):
    t = t.lower() 
    if(t=="á"):
        t="a"
    elif(t=="é"):
        t="e"
    elif(t=="í"):
        t="i"
    elif(t=="ó"):
        t="o"
    elif(t=="ú"):
        t="u"
    return t

#Inicializar texto sin tildes
textosintilde=""


#Recorrido de las lineas
for linea in textoinlines:
    #Recorrido de los caracteres en las lineas
    for caracter in linea:
        #Eliminacion de tildes de cada caracter recorrido
        textosintilde += quitartildes(caracter) 
#Printar texto sin tilde
print(textosintilde)

```

Resultado en nuevo archivo de texto resultado.txt.
Incluirá adicionalmente el número de líneas del fichero en la primer línea.
En la segunda el número de veces que aparece el árticulo el.

```python
import os

#Inicializacion número de filas
numrow=0
#Numero de filas
numrow=len(textoinlines)

#Abrir archivo
new= open ("resultado.txt", "w", encoding="utf-8")
#Escribir numero de filas y salto de linea
new.write(str(numrow)+ os.linesep)
#Escribir numero de artúclos y salto de linea
new.write(str(freq3)+ os.linesep)
#Escribir texto sin tilde 
new.write(textosintilde)
#Cerrar archivo
new.close()  
```

# Parte 2

Cargar archivo poblacionMunicipios.csv respetando tildes y carácteres especiales.

```python
#Importar libreria pandas
import pandas as pd
#Lectura de archivo csv 
dfpoblacion=pd.read_csv('poblacionMunicipios.csv')
```

Eliminación de las filas correspondientes a municipios sin población.

```python
#Guardar solo las filas que no tienen en la columna Poblacion 0
dfpoblacion=dfpoblacion[dfpoblacion["Poblacion"]!=0]
#Mostrar dicho csv
display(dfpoblacion)
```

Suma de la columna población.

```python
#Guardar y printar la suma de las poblaciones
sumpoblacion=dfpoblacion['Poblacion'].sum()
print(sumpoblacion)
```

Nuevo DataFrame, primera columna: provincias, segunda: total de habitantes por provincia, 
tercera: desviación, cuarta: número total de municipios por provincia.

```python
import numpy as np
#Suma de las poblaciones por provincia
sumapoblacionporprovincia=(dfpoblacion.groupby("Provincia")['Poblacion'].sum())

#Conversion a DataFrame
sumapoblacionporprovincia = pd.DataFrame(sumapoblacionporprovincia)
#Renombre de la columna Poblacion
sumapoblacionporprovincia=sumapoblacionporprovincia.rename(columns = {'Poblacion':'Población'})

#Desviacion de la Poblacion
desviacion=dfpoblacion.groupby("Provincia")['Poblacion'].std()
#Conversion a DataFrame
desviacion=pd.DataFrame(desviacion)
#Renombre de la columna Poblacion
desviacion=desviacion.rename(columns = {'Poblacion':'Desviacion'})

#Conteo de mmunicipios por población
nummunicipios=dfpoblacion.groupby("Provincia")['Municipio'].count()
#Conversion a DataFrame
nummunicipios=pd.DataFrame(nummunicipios)
#Renombre de la columna municipios
nummunicipios=nummunicipios.rename(columns={'Municipio':'Num de Municipios'})

#Lista de DataFrames
conjuntodeagrupaciones=[sumapoblacionporprovincia,desviacion,nummunicipios]

#DataFrame concadenando los distintos DataFrames 
concadenacion = pd.concat(conjuntodeagrupaciones,axis=1)
#Mostrar DataFrame solicitado por el enunciado
display(concadenacion)
```

Carga el archivo CP_Municipios.csv

```python
df_municipios=pd.read_csv('CP_Municipios.csv')
```

Cruce de datos.

```python
#Cruce del archivo población con el archivo municipios, sin perder datos, outer
merge1=pd.merge(dfpoblacion, df_municipios, how='outer', on=["CodMunicipio","CodProvincia"])
display(merge1)
```

Crear CSV de nombre faltan con los CP que no tienen asociados población o provincia.

```python
#Dataframe con los CP sin Provincia
nodataforCP=merge1[merge1['Provincia'].isnull()]
#Variable con el valor de dicho CP sin Provincia.
aux=nodataforCP["CP"].values
#Cambio de nombre en columnas e indices
dfAux=pd.DataFrame(data=aux,columns=["CP Sin Municipio"],index=[""])
dfAux.to_csv('faltan.csv', sep=',')
#Abrir y printar nuevo csv
with open ("faltan.csv", "r") as faltanfile:
    faltanfileloaded = faltanfile.read()
    print(faltanfileloaded)
```

Agrupacion de informacion en un solo DataFrame.
Primera columna: CP.
Segunda columna: Numero de municipios asociados a dicho CP.
Tercera columna: Suma de la problación en ese CP.

Cuarta: Nombre de la provincia.

```python
#Unir
agrup=pd.merge(dfpoblacion, df_municipios, how='outer', on=["CodMunicipio","CodProvincia"])
#Eliminacion de las columnas
agrup_reduced=agrup.drop(['CodMunicipio', 'Municipio_x','Municipio_y','CodProvincia','Poblacion',], axis=1)
#Eliminacion de las filas duplicadas
agrup_reduced=agrup_reduced.drop_duplicates()

#Cuenta del numero de municipios para el mismo CP
numMunicip=df_municipios.groupby("CP")['Municipio'].count()

#Suma de la población para CP
poblacionporCP=agrup.groupby("CP")['Poblacion'].sum()
#display(len(poblacionporCP))

#Escritura de las distintas columnas a partir de los datos de los DataFrames creados anteriormente
dataconjunto=pd.DataFrame({"CP":numMunicip.index,"NumMunicip":numMunicip.values,"Poblacion":poblacionporCP.values})

#Fusion tomando solo los datos contenidos en ambos data frame de:
#dataconjunto: NumMunicip y Poblacion
#agrup_reduced: Provincia
mergeresult=pd.merge(dataconjunto, agrup_reduced, how='inner', on=["CP",])
mergeresult.to_csv('comp.csv', sep=',')

#Muestra mergeresult
display(mergeresult)
```
